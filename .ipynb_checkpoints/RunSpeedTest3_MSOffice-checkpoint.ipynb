```python
import os
import re
import time
import random
import subprocess
from datetime import datetime
```

```python
def GetSpeeds():
    try:
        out = subprocess.Popen(['speedtest']
                               , stdout=subprocess.PIPE
                               , stderr=subprocess.STDOUT)
        stout, stderr = out.communicate()
    except:
        pass
    
    if stderr:
        pass
    else:
        pattern = '\d+.\d+ Mbps'
        download, upload = re.findall(pattern, str(stout))
    
    return(download, upload)
```

```python
datetimes = []
speedlist = []

for i in range(10):
    
    # wait for up to 5 minutes
    secs = random.randint(1, 60*5)
    time.sleep(secs)
    
    # get speeds
    speedlist.append(GetSpeeds())
    
    # get dates and times
    now = time.time()
    dt_obj = datetime.fromtimestamp(now)
    dd, tt = str(dt_obj).split(' ')
    tt = tt.split('.')[0]  
    
    datetimes.append((dd,tt))
```

```python
results = {'speeds':speedlist, 'times':datetimes}
```

```python
def ExtractSpeed(dic, direction):
    
    if direction == 'download':
        L = []
        for i in dic['speeds']:
            L.append(float(i[0].split(' ')[0])) 
        return(L)
    else:
        L = []
        for i in dic['speeds']:
            L.append(float(i[1].split(' ')[0]))
        return(L) 
    
def ExtractTime(dic):
    
    L = []
    for i in dic['times']:
        L.append(i[1])
    return(L)
```

```python
Y1 = ExtractSpeed(results, 'download')
Y2 = ExtractSpeed(results, 'upload')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = [10, 5]

labs = ExtractTime(results)
plt.xticks(range(10), labs, rotation='vertical')
plt.plot(Y1)

plt.ylabel('Download Speeds')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [10, 5]

labs = ExtractTime(results)
plt.xticks(range(10), labs, rotation='vertical')
plt.plot(Y2)

plt.ylabel('Upload Speeds')
plt.show()
```
